# Q3 The coin game with DP (20 marks)

Alice and Bob are playing a game using a bunch of coins.
The players pick several coins out of the bunch in turn.
Each time a player is allowed to pick 1, 2 or 4 coins, and the player that gets the last coin is the winner. Assume that both players are very smart and he/she will try his/her best to work out a strategy to win the game.
We suppose that Alice is always the first player to pick.
For example, if there are 2 coins, Alice will definitely pick 2 coins and win.
If there are 3 coins and Alice is still the first player to pick, no matter she picks 1 or 2 coins, Bob will get the last coin and win the game.
Suppose Alice plays first.
Given the number of coins, you are required to write a function *coingame* to calculate the winner of the game, and calculate how many different strategies there are for he/she to win the game.
You must use *dynamic programming* to solve the problem.

Design and implement a DP agorithm that solves the coin game, returning the name of the winner and the number of ways to win in a tuple (see unit tests).

```python
def findnonplayer(player):
    if player == "Alice":
        return "Bob"
    else:
        return "Alice"

def coingame_dp(n):
    #we prepare memoisation
    mem = [[None for coin in range(max(4,n)+1)] for role in ["playing", "nonplaying"]]
    #mem[0][i] tells us the number of ways to win with i coins
    #for the player who is playing now
    #mem[1][i] tells us the same, but for the player who is not playing
    
    #we initialise the base cases
    mem[0][1] = 1
    mem[1][1] = 0
    mem[0][2] = 1
    mem[1][2] = 0
    mem[0][3] = 0
    mem[1][3] = 2
    mem[0][4] = 3
    mem[1][4] = 0
    #we start the recursion and print the result
    return coingame(n, mem, "Alice")
    
def coingame(n, mem, player):
    nonplayer = findnonplayer(player)
    for moves in range(5,n+1):
        playerwaystowins = 0
        nonplayerwaystowins = 0
        for choice in [1, 2, 4]:
            #After making the choice from 1,2,4 the current player ('Alice') becomes the non player
            #Similarly the current non player ('Bob') becomes the next player with remaining coins 
            playerwaystowins += mem[1][moves-choice]
            nonplayerwaystowins += mem[0][moves-choice]
        if playerwaystowins > 0: #the current player can win
            mem[0][moves] = playerwaystowins
            mem[1][moves] = 0 # since if there is a way the current player can win the he/she will definitly take that chance
        else:
            mem[0][moves] = 0
            mem[1][moves] = nonplayerwaystowins

    if mem[0][n] > 0:
        return player,mem[0][n]
    return nonplayer, mem[1][n]
```
